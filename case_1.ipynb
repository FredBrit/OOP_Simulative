Extraction и Sale

```python
import csv
class Extraction:
  def __init__(self):
    ...

  @staticmethod
  def from_csv(file_path):
    file_path=file_path
    res=[]
    with open(file_path, 'r') as csvfile:
      csvreader = csv.reader(csvfile)
      for row in csvreader:
        res.append(row)
    return res

class Sale(Extraction):

  def __init__(self,file_path,id,date,amount,product):

    self.res = self.from_csv(file_path)
    self.id=id
    self.date=date
    self.amount=amount
    self.product=product

    self.sale=[]
    for el in self.res:
      if el[0]==id and el[1]==date and el[2]==amount and el[3]==product:
        self.sale.append(el)

    self.sale=self.sale[0]


  
  def get_id(self):
    return self.id

  def set_id(self,id):
    self.id=id

  def get_date(self):
    return self.date

  def set_date(self,date):
    self.date=date

  def get_amount(self):
    return self.amount

  def set_amount(self,amount):
    self.amount=amount

  def get_product(self):
    return self.product

  def set_product(self,product):
    self.product=product
#Работает если вызывать так, но проверку не проходит, так как объект класса состоит из списков и к ним не применить метод, буду признателен, если скинете образцовый код
#e=Extraction.from_csv('sales_data.csv')
#s=Sale('sales_data.csv','1','2023-04-25','35.21','Product 6')
#s.get_id()

```

Transformation и Analysis

```python
class Transformation(Extraction):

  def __init__(self,file_path):

    self.res = self.from_csv(file_path)

  def filter_by_date(self,start_date,end_date):
    self.start_date=start_date
    self.end_date=end_date
    filtered_sales_by_dt=[]
    for el in self.res:
      if el[1]>=start_date and el[1]<=end_date:
        filtered_sales_by_dt.append(el)

    return filtered_sales_by_dt


  def filter_by_amount(self,min_amount,max_amount):
    self.min_amount=min_amount
    self.max_amount=max_amount
    filtered_sales_by_amount=[]
    for el in self.res[1:]:
      if float(el[2])>=min_amount and float(el[2])<=max_amount:
        filtered_sales_by_amount.append(el)

    return filtered_sales_by_amount



#t=Transformation('sales_data.csv')
#t.filter_by_date('2023-04-25','2023-04-26')
#t.filter_by_amount(35.21,40)


class Analysis(Extraction):

  def __init__(self,file_path):

    self.res = self.from_csv(file_path)

  def calculate_total_sales(self):
    amounts=[]
    for el in self.res[1:]:
        amounts.append(float(el[2]))
    return round(sum(amounts),2)

  def calculate_average_sales(self):
    amounts=[]
    for el in self.res[1:]:
        amounts.append(float(el[2]))
    return round(sum(amounts)/len(amounts),2)

#a=Analysis('sales_data.csv')
#a.calculate_total_sales()
#a.calculate_average_sales()

#Работает, но также не проходит проверку от платформы
```

Loading

```python
class Loading(Extraction):

  def __init__(self,file_path):

    self.res = self.from_csv(file_path)

  def to_csv(self):

    file_name = 'filtered_sales.csv'
    with open(file_name, mode='w', newline='') as csvfile:
      csvwriter = csv.writer(csvfile)
      for el in self.res:
        csvwriter.writerow([el])

#l=Loading('sales_data.csv')
#l.to_csv()

#Также не проходит проверку, но работает
```
